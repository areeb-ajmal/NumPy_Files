```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("=== Zomato sales analysis ===")
print(f"\n Sales data shape {sales_data.shape}")
print(f"\n Sample data of first 3 resturents : {sales_data[0 : 3]}")
```

```python
#Total sales per YEAR

yearly_total = np.sum(sales_data[ : , 1 : ], axis=0)
print(f"Total Sales per YEAR : \n {yearly_total}")

#Minium sales per YEAR

min_sales = np.min(sales_data[ : , 1 : ], axis=1)
print(f"Minimum sales per YEAR : \n {min_sales}")

#Max sales per YEAR

max_sales = np.max(sales_data[ : , 1 : ], axis=0)
print(f"Maximum sales per YEAR : \n {max_sales}")

#Average sales per resturant

avg_sales = np.mean(sales_data[ : , 1 : ], axis=1)
print(f"Average sales per RESTURANT : \n {avg_sales}")



```

```python
cumsum = np.cumsum(sales_data[ : ,1 : ], axis=1)
print(f"Cumulative sum : \n {cumsum}")

plt.figure(figsize=(8, 6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average of sales per resturant")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()

```

### Vector Operations

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 9, 10])

print(f"Vector addition : \n {vector1 + vector2}")
print(f"Multiplication of vector : \n {vector1 * vector2}")
print(f"Dot product of vectors : \n {np.dot(vector1, vector2)}")

angle = np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2))
print(f"Angle of vectors : \n {angle}°")
```

```python
restaurant_types = np.array(['biryani', 'chinese', 'pizza', 'burger', 'cafe'])
vectorized_upper = np.vectorize(str.upper)

print(f"Vectorized upper : \n {vectorized_upper(restaurant_types)}")
```
