```python
import numpy as np
```

```python
array1 = np.array([[1, 2, 3],
                   [4, 5, 6]])

array2 = np.random.rand(3, 3)
array3 = np.zeros((4, 4))
```

```python
np.save("array1.npy", array1)
np.save("array2.npy", array2)
np.save("array3.npy", array3)
```

```python
array1_loaded = np.load("array1.npy")
print(f"Array1 : \n {array1_loaded}")
array2_loaded = np.load("array2.npy")
print(f"Array2 : \n {array2_loaded}")
array3_loaded = np.load("array3.npy")
print(f"Array3 : \n {array3_loaded}")


```
